```python
import torch
import torchvision
import torchvision.transforms as transforms
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

# Assume that we are on a CUDA machine, then this should print a CUDA device:

print(device)
```

```python
transform = transforms.Compose([
    transforms.Lambda(lambda x: x.convert('RGB') if x.mode != 'RGB' else x),
    transforms.Resize((224, 224)),  # Thay đổi kích thước hình ảnh
    transforms.ToTensor(),  # Chuyển đổi PIL Image sang Tensor
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))  # Chuẩn hóa
])


# Tải dữ liệu từ thư mục của bạn
trainset = torchvision.datasets.ImageFolder(root='/kaggle/input/dataset/dataset/train', transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=4, shuffle=True, num_workers=2)

testset = torchvision.datasets.ImageFolder(root='/kaggle/input/dataset/dataset/test', transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=4, shuffle=False, num_workers=2)

# Định nghĩa các lớp trong bộ dữ liệu của bạn
classes = ('alligator', 'koi', 'shark')
import matplotlib.pyplot as plt
import numpy as np
import torchvision

# Định nghĩa các lớp tương ứng trong bộ dữ liệu mới
classes = ('alligator', 'koi', 'shark')

def imshow(img):
    img = img / 2 + 0.5     # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))  # chuyển từ (C, H, W) sang (H, W, C)
    plt.show()

# Lấy một số hình ảnh ngẫu nhiên từ tập dữ liệu huấn luyện
dataiter = iter(trainloader)
images, labels = next(dataiter)

# Hiển thị hình ảnh
imshow(torchvision.utils.make_grid(images))

# In nhãn tương ứng
print(' '.join(f'{classes[labels[j]]:5s}' for j in range(len(labels))))  # Sửa đổi để phù hợp với số lượng hình ảnh thực tế trong một batch
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F


class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, 7, padding = 3, stride = 1)

        self.conv2_5 = nn.Conv2d(32, 64, 5, padding = 2, stride = 2)
        self.conv3_3 = nn.Conv2d(32, 64, 3, padding = 1, stride=2)

        self.conv4_5 = nn.Conv2d(64, 96, 5, padding = 2, stride=1)
        self.conv5_3 = nn.Conv2d(64, 96, 3, padding = 1, stride=1)

        self.conv6_3 = nn.Conv2d(96, 128, 3, padding = 1, stride=1)
        self.conv7_3 = nn.Conv2d(96, 128, 3, padding = 1, stride = 1)
        self.conv8_5 = nn.Conv2d(128, 160, 5, padding = 2, stride = 1)

        self.conv9_3 = nn.Conv2d(128, 160, 3, padding = 1, stride = 1)

        self.conv10_5 = nn.Conv2d(160, 192, 5, padding = 2, stride = 1)
        self.conv11_3 = nn.Conv2d(160, 192, 3, padding = 1, stride = 1)
        self.conv12_5 = nn.Conv2d(192, 224, 5, padding = 2, stride = 1)
        self.conv13_3 = nn.Conv2d(224, 256, 3, padding = 1, stride = 1)
        self.conv14_5 = nn.Conv2d(256, 288, 5, padding = 2, stride = 1)
        self.conv15_3 = nn.Conv2d(288, 320, 3, padding = 1, stride = 1)
        self.conv16_5 = nn.Conv2d(576, 608, 5, padding = 2, stride = 2)
        self.conv17_5 = nn.Conv2d(608, 640, 5, padding = 2, stride = 2)
        self.conv18_5 = nn.Conv2d(320, 384, 5, padding = 2, stride = 2)
        self.conv19_3 = nn.Conv2d(224, 384, 3, padding = 1, stride = 2)
        self.conv20_5 = nn.Conv2d(384, 512, 5, padding = 2, stride = 1)

        self.maxpooling=nn.MaxPool2d(2)
        self.maxpooling1=nn.MaxPool2d(14)
        self.avg=nn.AvgPool2d(2)

        self.avg2= nn.AvgPool2d(7)


        self.fc1 = nn.Linear(896, 512)
        self.fc2 = nn.Linear(512, 128)
        self.fc3 = nn.Linear(128, 3)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        
        x1_1 = F.relu(self.conv2_5(x))
        x1_2 = F.relu(self.conv3_3(x))

        x1_1 = F.relu(self.conv4_5(x1_1))
        x1_2 = F.relu(self.conv5_3(x1_2))

        x1_1 = F.relu(self.conv6_3(x1_1))
        x1_2 = F.relu(self.conv7_3(x1_2))
        
        x2_1 = F.relu(self.conv8_5(x1_1))
        x2_2 = F.relu(self.conv8_5(x1_1))
        x2_1 = self.avg(x2_1)
        x2_2 = self.maxpooling(x2_2)
        x4 = F.relu(x2_1+x2_2)
        x3=torch.cat((x1_1, x1_2), dim=1)

        x3_1 = F.relu(self.conv9_3(x1_2))
        x3_2 = F.relu(self.conv9_3(x1_2))
        x5 = torch.cat((x3_1, x3_2), dim=1)
        x5 = self.maxpooling(x5)
        x6_1=F.relu(self.conv10_5(x4))
        x6_2=F.relu(self.conv11_3(x4))
        x6_1 = F.relu(self.conv12_5(x6_1))
        x6_1 = F.relu(self.conv13_3(x6_1))
        x6_1 = F.relu(self.conv14_5(x6_1))
        x6_1 = F.relu(self.conv15_3(x6_1))
        x7 = F.relu(x5+x6_1)
        
        x6_3 = F.relu(self.conv10_5(x4))
        x8_1 = self.avg(x6_3)
        x8_2 = self.maxpooling(x6_3)
        x9 = F.relu(x6_2+x6_3)
        x9 = F.relu(self.conv12_5(x9))
        x10 = torch.cat((x8_1, x8_2), dim=1)
        x7 = F.relu(self.conv18_5(x7))
        x11= F.relu(x10 + x7)
        x11 = F.relu(self.conv20_5(x11))
        x9 = F.relu(self.conv19_3(x9))
        x12 = torch.cat((x9, x11), dim=1)
        x12 = self.avg(x12)
        x12 = self.maxpooling1(x12)
        x12 = x12.view(-1, 896)
        x12 = (self.fc1(x12))
        x12 = (self.fc2(x12))
        x12 = self.fc3(x12)
        return x12
        


net = Net()
net = net.to(device)
from torchsummary import summary
summary(net,(3, 224, 224))
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
```

```python
import warnings

# Bỏ qua cảnh báo cụ thể về hình ảnh có độ trong suốt
warnings.filterwarnings("ignore", message="Palette images with Transparency expressed in bytes should be converted to RGBA images")
```

```python

for epoch in range(20):  # loop over the dataset multiple times

    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        # get the inputs
        inputs, labels = data
        inputs, labels = inputs.to(device), labels.to(device)

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        # print statistics
        running_loss += loss.item()
        if i % 50 == 49:   
            print('[%d, %5d] loss: %.3f' %
                  (epoch + 1, i + 1, running_loss / 50))
            running_loss = 0.0

print('Finished Training')
```

```python
correct = 0
total = 0
with torch.no_grad():
    for data in testloader:
        images, labels = data
        images, labels = images.to(device), labels.to(device)
        outputs = net(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print('Accuracy of the network on the 10000 test images: %d %%' % (
    100 * correct / total))
```

```python
class_correct = list(0. for i in range(3))
class_total = list(0. for i in range(3))
with torch.no_grad():
    for data in testloader:
        images, labels = data
        images, labels = images.to(device), labels.to(device)
        outputs = net(images)
        _, predicted = torch.max(outputs, 1)
        c = (predicted == labels).squeeze()
        for i in range(3):
            label = labels[i]
            class_correct[label] += c[i].item()
            class_total[label] += 1


for i in range(3):
    print('Accuracy of %5s : %2d %%' % (
        classes[i], 100 * class_correct[i] / class_total[i]))
```
